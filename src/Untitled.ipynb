```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from os import listdir

bin_path = '/Users/harihara/Mount/projects/refining_bins_with_assembly_graphs/Scaffold_Coverage_binning/delinked_scaffold_bins/'
coverage_path = '/Users/harihara/Research-Activities/Data/Binnacle-Op/Scaffold_Coverage_After_Delinking/'
```

```python
def Is_Scaffold(Scaffold_ID):
    if 'cc_' in Scaffold_ID:
        return 'Scaffold'
    else:
        return 'Contig'
    
def Get_Scaffolds(sample):
    df = pd.read_csv(bin_path+sample+'_delinked_cc_bins/'+sample+'_delinked_cc', 
                     sep = '\t', names = ['Connected_Component_After_Delinking', 'Bin_ID'])
    df['Type'] = df['Connected_Component_After_Delinking'].apply(Is_Scaffold)
    df['Connected_Component_After_Delinking'] = df['Connected_Component_After_Delinking'].str.replace("cc_","")
    return df

def Load_Coverage_Summary(sample):
    df_cov = pd.read_csv(coverage_path+sample+'_Coverages.csv_Summary.csv', index_col = ['Connected_Component'])
    return df_cov
```

```python
df_bins = Get_Scaffolds('SRS104485')
```

```python
df_cov_summary = Load_Coverage_Summary('SRS104485')
```
